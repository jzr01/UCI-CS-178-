## Plotting the Data

There are a 100001 ways to visualize the data, but in this course we'll use the simple built in [matplotlib.pyplot](https://matplotlib.org/api/pyplot_api.html) package. The nice thing about it that it can be used in a very simple way but also you can do crazy things with it.

First thing first, if you are using Jupyter Notebooks you need to allow inline plotting. Just run the next code cell.

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

The most simple type of plotting is plotting a line. Let's plot the first two columns of the iris data using the first one as the x-axis and the second as the y-axis.

```python
plt.plot(iris[:, 0], iris[:, 1])
plt.show() # You don't have to have this in the notebooks, but it's goot practice as it cleans the canvas.
```

This plot doesn't make much sense, but that's because this is not the right way to plot these data. A more sensible way is a scatter plot:

```python
plt.scatter(iris[:, 0], iris[:, 1])
plt.show() # You don't have to have this in the notebooks, but it's good practice as it cleans the canvas.
```

Now let's plot the scatter using the same two columns and the last one as the class (which you will have to do in your HW assignment).

Notice that I also added a lot more stuff in the scatter method. There are 100001 options in the pyplot library for each method, you can see them in the documentations and plotting examples online.

```python
plt.scatter(iris[:, 0], iris[:, 1], s=130, c=iris[:, -1], alpha=0.75)
plt.show()
```

Or you can use a different technique, which allows you to control more things -- plot each class seperately (either in a loop or, if you know in advance what you want to plot, manually)

```python
colors = ['blue', 'green', 'red']

for i, c in enumerate(np.unique(iris[:, -1])):
    mask = np.where(iris[:, -1] == c)[0]   # Finding the right points
    plt.scatter(iris[mask, 0], iris[mask, 1], s=120, c=colors[i], alpha=0.75, label='class %d' % i)

plt.legend()
plt.show()
```

There are a lot of different ways to plot things, some are more pretty, some are less. At the end of the day you should plot things in the way that is most supporting your results or help you find what you're looking for in the data.

## More plotting

A few features of the plotting system are pretty useful: 
* Setting plot sizes
* Multiple plots per cell

You can make your plots short and wide:

```python
plt.figure( figsize = (8,3) )
plt.plot(np.random.rand(30),'bo')
plt.show()
```

or tall and skinny:

```python
plt.figure( figsize = (2,5) )
plt.plot(np.random.rand(4),'r-')
plt.show()
```

and you can put multiple plots together in a variety of ways (see matplotlib documentation).  The simplest is to use "subplots", which define a collection of "axes" with sort-of similar interfaces to the main plot functions:

```python
fig, ax = plt.subplots(3,2)
ax[0][0].plot( np.random.rand(10), 'gx')  # a bunch of green x's
ax[1][0].plot( np.random.rand(7), 'm-')   # a random magenta line
ax[2][1].plot( [0],[0], 'yo', ms=30)      # giant yellow dot at (0,0)
```

## Iterative plotting

Sometimes, we will be running something for a long time, and want to output information / plot it before the end to get a sense of what's happening.  In this case, you can plot, and then clear the jupyter display before plotting again, in order to replace the first output with the new output when it is available:

```python
import IPython.display as display   # for manipulating the jupyter display
import time                         # for pretending to do something slow

y = iris[:,-1]
for c in range(int(y.max())):   # run through the classes
    # plot all the data for classes 0...c
    plt.scatter(iris[y<=c, 0], iris[y<=c, 1], s=130, c=iris[y<=c, -1], alpha=0.75)
    plt.show()
    # now pretend we have to do some work:
    time.sleep(5)
    display.clear_output()

# re-plot at the end (since we just cleared the display again)
plt.scatter(iris[:, 0], iris[:, 1], s=130, c=iris[:, -1], alpha=0.75)    
plt.show()
```

This is particularly useful for plotting, e.g., model performance over time as we train.  We can simulate such a plotting as:

```python
f = np.zeros(30)         # I like to pre-allocate storage
for i in range(len(f)):
    time.sleep(0.5)       # pretend to do something to improve our model here...
    f[i] = np.exp(-.1*i)  # pretend to compute the current performance of our model
    
    display.clear_output()
    plt.plot(range(i+1),f[:i+1],'bx-')
    plt.axis([0,30,0,1])  # see note below
    plt.show()
```

You'll notice:
* If you don't have enough stuff below this cell, Jupyter will annoyingly shift you around as it redisplays
* If you don't set the plot range manually, it will auto-set the plot each time.  I like the appearance to be consistent from step to step, so I set it manually with "axis" or "xrange"
